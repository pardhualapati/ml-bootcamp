#STEP 1:IMPORT DATA

```python
import pandas as pd
import numpy as np
data = pd.read_csv("50_Startups_MLR.csv")
print(data)
```

#STEP 2: REPLACE ZERO WITH NAN

```python
data["Marketing Spend"] = data["Marketing Spend"].replace(0,np.nan)
data["R&D Spend"] = data["R&D Spend"].replace(0,np.nan)
data["Administration"] = data["Administration"].replace(0,np.nan)
```

```python
data
```

#STEP 3: SEPERATE I/P & O/P

---



```python
#step-2: Now we have to seperate the both input and output varibles
#Independent Varables
X = data.iloc[:,0:-1].values
Y=data.iloc[:,-1].values
```

```python
X
```

#STEP 4 : HANDLING MISSING VALUES

```python
#Step-3:TALK ABOUT MISSING DATA
from sklearn.impute import SimpleImputer
import numpy as np
imputer= SimpleImputer(missing_values = np.nan,strategy="mean")
X[:,0:3] = imputer.fit_transform(X[:,0:3])
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
composer = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X =composer.fit_transform(X)
print(X)
```

```python
# Encode the dependent variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X[:,0:3] = sc.fit_transform(X[:,0:3])
```

```python
X
```

#Step 6: Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train ,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
regressor.predict(X_test)
```

```python
from flask import Flask, render_template, request
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)
@app.route('/')
def home():
  return render_template('index.html')

@app.route('/predict', methods=['POST'])
def collectData():
  rad = float(request.form['rad'])
  admin = float(request.form['admin'])
  mar = float(request.form['mar'])
  city = request.form['City']
  k = []
  if city == 'New York':
    k=[0,0,1]
  elif city == 'California':
    k=[1,0,0]
  elif city == 'Florida':
    k=[0,1,0]
  else:
    k=[0,0,0]
  k.append(rad)
  k.append(admin)
  k.append(mar)
  pre = regressor.predict([k])[0]
  return render_template('result.html', predict=pre)

if __name__ == "__main__":
  app.run()
```
